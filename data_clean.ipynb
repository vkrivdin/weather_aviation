```python
import pandas as pd
```

```python
df = pd.read_csv('flight_data.csv')
```

```python
df.head()
```

```python
df['departure_code'] = df['Departure_Airport'].str.split(' Airport', n=1, expand= True)[0]

df
```

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
df['arrival_code'] = df['Arrival_Airport'].str.split(' Airport', n=1, expand= True)[0]

df = df.drop(columns=['Arrival_Airport','Departure_Airport', 'Date', 'Time'], inplace=False)

df.index = df['Flight_ID']

df.info()
```

```python
df.to_csv('data_clean.csv')
```
